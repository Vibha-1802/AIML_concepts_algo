```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
dataset = pd.read_csv("KNN.csv")
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
print(X.head())
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.10)
classifier = KNeighborsClassifier(n_neighbors=5).fit(Xtrain,ytrain)
ypred=classifier.predict(Xtest)
i=0
print("\n----------------------------------------------------------")
print('Original Label','Predicted Label','Correct/Wrong',sep="\t")
print("------------------------------------------------------------")
for label in ytest:
    print(label,ypred[i],sep="\t",end="\t")
    if(label == ypred[i]):
       print("Correct")
    else :
       print("Wrong")
    i = i+1
print("-------------------------------------------------------------")
print("\nConfusion Matrix:\n",metrics.confusion_matrix(ytest,ypred))
print("-------------------------------------------------------------")
print("\nConfusion Matrix:\n",metrics.classification_report(ytest,ypred))
print("-------------------------------------------------------------")
print("\nConfusion Matrix:\n",metrics.accuracy_score(ytest,ypred))
```
